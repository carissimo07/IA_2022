```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("datosConcatenadossinCI.csv", index_col = 0)
```

```python
df
```

```python
df1 = df[ (df["Cod.Curso"] == 1 ) & ( (df["Cod.Car.Sec"] == "ECA-PLS13 ") | (df["Cod.Car.Sec"] == "ECA-PLS09 ") ) ]
df1 = df1.drop_duplicates()
df1
```

```python
df2 = df[ (df["Cod.Curso"] == 4) & (  (df["Cod.Car.Sec"] == "ECA-PLS13 ") | (df["Cod.Car.Sec"] == "ECA-PLS09 ") ) & (df["Aprobado"] == "S")]
df2 = df2.drop_duplicates()
df2
```

```python
#Algoritmo para saber si cursan las 6 materias de 1er Semestre
lista1 = df1['id_anony'].tolist()
lista2 = []
for i in range(len(lista1)):
  if lista1.count(lista1[i]) == 6:
    lista2.append(lista1[i])
data1 = pd.DataFrame(lista2, columns = ['id_anony'])
data1 = data1.drop_duplicates(subset=['id_anony'])
data1
```

```python
#Algoritmo para saber si pasaron las 5 materias de 4to Semestre
lista3 = df2['id_anony'].tolist()
lista4 = []
for i in range(len(lista3)):
  if lista3.count(lista3[i]) == 5:
    lista4.append(lista3[i])
data2 = pd.DataFrame(lista4, columns = ['id_anony'])
data2 = data2.drop_duplicates(subset=['id_anony'])
data2
```

```python
entrada = df1[['Asignatura' , 'Anho','id_anony']]
entrada = entrada.drop_duplicates(subset=['id_anony'] , keep = 'first')
entrada = pd.merge(entrada, data1, on = 'id_anony')
entrada
```

```python
salida = df2[['Asignatura','Anho','id_anony']]
salida = salida.drop_duplicates(subset=['id_anony'], keep = 'last')
salida = pd.merge(salida,data2, on = 'id_anony')
salida
```

```python
result = pd.merge(entrada, salida, on = 'id_anony')
result
```
